```python
import numpy as np
import rasterio as rio
from rasterio.plot import show
import matplotlib.pyplot as plt
```

```python
filename = "/content/drive/MyDrive/Classroom/Curso de Manejo de drones y procesamiento de imágenes con Inteligencia Artificial/DEM_LaCalera_USGS.tif"
```

```python
dem = rio.open (filename)
dem_array = dem.read (1).astype ('float64')
```

```python
fig,ax = plt.subplots(1, figsize=(12, 12))
show(dem_array, cmap='gist_rainbow_r', ax=ax)
show(dem_array, contour=True, ax=ax, linewidths=0.5)
plt.axis('on')
plt.show()
```

```python
dem_array[dem_array == filename['nodata']] = np.nan
```

```python
plt.hist(dem_array)
```

```python
import numpy as np
import rasterio as rio
import richdem as rd
```

```python
dem = rio.open (filename)
dem_array = dem.read (1).astype('float64')
dem_richdem = rd.rdarray (dem_array, no_data=-9999)
```

creado el RichDEM podemos analizar y crear visualizaciones del DEM 

```python
fig = rd.rdShow(dem_richdem, axes=False, cmap='jet', figsize=(16, 16))
```

```python
dem_filled = rd.FillDepressions(dem_richdem, in_place=False)
dem_filled_fig = rd.rdShow(dem_filled, ignore_colours= [0], axes=False, cmap='jet', figsize=(16, 16))
```
